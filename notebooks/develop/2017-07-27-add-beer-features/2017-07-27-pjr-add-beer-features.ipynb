# Load Data

```python
datapath = './data'
# load beer_eda.csv
bfile = os.path.join(datapath, 'beer_eda.csv')
beer = pd.read_csv(bfile)
beer.fname = 'beer-data'
```

```python
del beer['Unnamed: 0']
```

```python
beer = beer.rename(columns={'id': 'beer_id', 'post_code': 'postal_code_id'})
```

```python
beer.head()
```

```python
beer.zip_code = beer.zip_code.apply(lambda x: str(x).zfill(5)).astype(str)
beer.zip_code = beer.zip_code.apply(lambda x: str(x).strip('.0'))
```
