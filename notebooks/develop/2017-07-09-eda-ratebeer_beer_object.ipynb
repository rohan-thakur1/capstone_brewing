```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Data Exploration 

```python
df_beer = pd.read_csv('../../data/rate-beer/beer_clean.csv')
df_beer.head(5)
```

```python
df_beer.describe()
```

### Univariate Analysis

```python
#Plot KDE Plots
%matplotlib inline
fig, axs = plt.subplots(3,3, figsize=(20,10))
#df_beer.weighted_avg.hist(ax=axs[0])
sns.kdeplot(df_beer.abv, shade = True,ax=axs[0][0])
sns.kdeplot(df_beer.calories, shade = True,ax=axs[0][1])
sns.kdeplot(df_beer.mean_rating, shade = True,ax=axs[0][2])
sns.kdeplot(df_beer.num_ratings, shade = True,ax=axs[1][0])
sns.kdeplot(df_beer.overall_rating, shade = True,ax=axs[1][1])
sns.kdeplot(df_beer.style_rating, shade = True,ax=axs[1][2])
sns.kdeplot(df_beer.weighted_avg, shade = True,ax=axs[2][0])
```

```python
#Plot boxplots
%matplotlib inline
fig, axs = plt.subplots(3,3, figsize=(20,10))
sns.boxplot(df_beer.abv, color = "green", orient = "v",ax=axs[0][0])
sns.boxplot(df_beer.calories, color = "green", orient = "v",ax=axs[0][1])
sns.boxplot(df_beer.mean_rating, color = "green", orient = "v",ax=axs[0][2])
sns.boxplot(df_beer.num_ratings, color = "green", orient = "v",ax=axs[1][0])
sns.boxplot(df_beer.overall_rating, color = "green", orient = "v",ax=axs[1][1])
sns.boxplot(df_beer.style_rating, color = "green", orient = "v",ax=axs[1][2])
sns.boxplot(df_beer.weighted_avg, color = "green", orient = "v",ax=axs[2][0])
```

*Takeaways*
1. Abv: Most beers have less than 10% alchohol with some high outliers, i.e., positive skew distribution.
2. Calories: Most beers have 200 cals (close to normal distributions with tails on either side)
3. Num_ratings: Most beers have less than 200 ratings, with very high outliers.
4. Style Ratings: Uninteresting as most get 100
5. Mean Rating: Weighted avg better
6. Overall Rating: Uninteresting as most have 100
7. Weighted Avg: Normal around 3.6-3.7 (not really "normal" since it can only go from 0-5)

### Bivariate Analysis

```python
#Plot scatterplots
%matplotlib inline
fig, axs = plt.subplots(2,3, figsize=(20,10))
sns.regplot("abv", "calories", data=df_beer, ax=axs[0][0])
sns.regplot("abv", "num_ratings", data=df_beer, ax=axs[0][1])
sns.regplot("abv", "weighted_avg", data=df_beer, ax=axs[0][2])
sns.regplot("calories", "num_ratings", data=df_beer, ax=axs[1][0])
sns.regplot("calories", "weighted_avg", data=df_beer, ax=axs[1][1])
sns.regplot("weighted_avg", "num_ratings", data=df_beer, ax=axs[1][2])
```

*Takeaways*
1. Calories seem to be closely related to % alc
2. No other very interesting correlations
